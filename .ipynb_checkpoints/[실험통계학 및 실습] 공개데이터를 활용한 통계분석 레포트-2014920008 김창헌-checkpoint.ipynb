<h1>

<h1 style="color:blue;">분석1: 도로 종류에 따른 사고건수에 차이가 있는가?</h1>

## 귀무가설 대립가설 설정

귀무가설 : 도로 종류에 따른 사고건수에 차이가 없다.  
대립가설 : 도로 종류에 따른 사고건수에 차이가 있다.

## 데이터 탐색

```r
#1. 데이터 입력
data = read.table("도로교통공단_도로종류별_시간대별_교통사고(2013).csv", header=T, sep=',')
str(data)
attach(data)
data2 = cbind(도로종류, 사고건수)
```

```r
data2 = data.frame(도로종류, 사고건수)
```

```r
options(repr.plot.width=5 , repr.plot.height=5)
plot(data2)
```

## 분산분석의 가정 성립 확인


#### (1)분산분석의 가정 확인 - 정규성


```r
options(repr.plot.width=5 , repr.plot.height=5)
accidents = aov(사고건수 ~ 도로종류, data=data)
aov_residuals = accidents$residuals
qqnorm(aov_residuals)
qqline(aov_residuals, col=2)
shapiro.test(x=aov_residuals) # p = 9.217e-06<0.05 정규성을 만족하지 않음.  
#-->--> 정규성을 만족하지 않고 1원이므로 
```

#### (2)분산분석의 가정 확인 - 등분산성
p = 5.174703e-08< 0.05이므로 등분산성 만족못함

```r
# install.packages("car")
require("car")
leveneTest(사고건수 ~ 도로종류) # p = 5.174703e-08< 0.05
```

## 데이터 분석 - 정규분포를 따르지 않고 factor의 수가 1개이므로 Kruskal -Walis검정

```r
# install.packages("nparcomp")
require(nparcomp)
kruskal.test(사고건수 ~ 도로종류, data=data2)#적어도 한 그룹이 다르다.
```

p-value = 1.084e-12 < 0.05 이므로 적어도 한 그룹이 차이가 있다. 

### 사후검정

```r
m_result = mctp(사고건수~도로종류, data=data2)# 사후검정
```

```r
summary(m_result)
```

유의미한 차이가 있는 그룹

```r
m_result$Analysis.Inf[m_result$Analysis.Inf$p.Value < 0.05 ,  ]
```

유의미한 차이가 없는 그룹

```r
m_result$Analysis.Inf[m_result$Analysis.Inf$p.Value > 0.05 ,  ]
```

<h2>결론 및 토의사항</h2>

    분석결과 귀무가설은 기각되어 대립가설이 채택되었다. 도로 종류에 따른 사고건수에 차이는 있는 것으로 밝혀졌고 특히 ‘시도’ ‘특별광역시도’에서 가장 사고가 많이 일어나는 것을 확인할 수 있었다.고속국도와 군도, 일반국도와 기타도로, 지방도와 기타도로, 지방도와 일반국도, 특별광역시와 시도 끼리는 사고건수의 유의미한 차이가 없었으며. 
가장 사고건수가 적은 곳은 고속국도였다.

<h1 style="color:blue;">분석2: 시간대에 따른 사고건수에 차이가 있는가?</h1>

귀무가설 : 시간대별 사고건수의 차이가 없다. 

대립가설 : 시간대별 사고건수의 차이가 있다. 

```r
data3 = data.frame(시간대, 사고건수)
str(data3)
```

```r
options(repr.plot.width=10 , repr.plot.height=10)
plot(data3)
```

## 분산분석의 가정 성립 확인


#### (1)분산분석의 가정 확인 - 정규성

```r
accidents = aov(사고건수 ~ 시간대, data=data3)
aov_residuals = accidents$residuals
qqnorm(aov_residuals)
qqline(aov_residuals, col=2)
shapiro.test(x=aov_residuals) # p = 2.37e-08<0.05 정규성을 만족하지 않음.  
#-->--> 정규성을 만족하지 않고 1원이므로 
```

#### (2)분산분석의 가정 확인 - 등분산성
p_value = 0.9603032 > 0.05 이므로 등분산성 만족 

```r
# install.packages("car")
require("car")
leveneTest(사고건수 ~ 시간대) # p = 0.9603032 > 0.05 
```

```r
require(nparcomp)
kruskal.test(사고건수 ~ 시간대, data=data3)#모든 그룹에서 차이가 없다. 
```

### kruskal.test 결과 p-value = 0.3563이므로 모든 그룹에서 유의미한 차이가 없다. 

```r
m_result = mctp(사고건수~시간대, data=data3)# 사후검정
```

```r
summary(m_result)
```

### 사후 검정 결과도 역시 차이가 없음을 알 수 있다. 

```r
m_result$Analysis.Inf[m_result$Analysis.Inf$p.Value < 0.05 ,  ]
```

```r
m_result$Analysis.Inf[m_result$Analysis.Inf$p.Value > 0.05 ,  ]
```

```r
library(ggplot2)
ggplot(data=data3, aes(x=시간대, y=사고건수))+
  geom_boxplot(alpha=0.1,aes(col=시간대,fill=시간대))+
  geom_jitter(alpha=0.3, aes(col=시간대),width=0.05)+theme_bw()+
  theme(text=element_text(size=14,face="bold"), plot.title=element_text(hjust=0.5, size=16,face="bold"))
```

정규성 테스트 결과 정규성을 만족하지 않았고 등분산성은 만족했다. 따라서 
kruskal.test를 진행하였는데 p-value = 0.3563이므로 모든 그룹에서 유의미한 차이가 없다는 결과가 나왔다. 시간대에 대한 유의미한 차이가 있을거라고 예상했는데 차이가 없다는 결과가 나와서 데이터를 살펴보았다. 


시간대별로 각 7개의 샘플이 존재했고 7개의 샘플은 각각 다른 도로 종류 값을 가지고 있었다. 시간대별로 사고건수의 차이가 있는 것으로 보이는데 어떻게 분석을 더 해 나가야할지 더 고민해 보아야 할 것 같다. 

# 반복측정 분산분석을 수행해야 했음을 깨달았다. 
이 데이터는 각 시간대별로 7번 반복측정을 한 결과라고 볼 수 있다. 

rmanova를 하려고 봤더니 정규성을 만족하지 않는다 그렇다면 어떻게 해야하는가? -->비모수 검정을 해야한다. 
즉 Ranked Transformed ANOVA를 해야한다. 

```r
require(ARTool)
require(ART)
require(phia)
```

```r
anova(result)
```

여전히 유의미한 차이가 없다고 나오는데.. 프리드만 테스트를 하라는 조언을 블로그에서 보았다. 
http://blog.naver.com/PostView.nhn?blogId=y4769&logNo=220377172228&beginTime=0&jumpingVid=&from=search&redirect=Log&widgetTypeCall=true
1. 정규성을 만족하지 못하면 모수통계는 적용할 수 없다.

2. RM ANOVA에서 정규성을 만족하지 못하면 Friedman test 를 시행한다.
[출처] [SPSS 22] 반복측정 분산분석(Repeated Measures ANOVA)|작성자 도시마법사

```r
data$시간대
```

```r
library(rcompanion)
library(multcompView)
require(PMCMR)
friedman.test(data$사고건수,data$도로종류, data$시간대)
Post = posthoc.friedman.nemenyi.test(data$사고건수,data$도로종류, data$시간대)  # p-value = 3.174e-07 < 0.05 로 유의미한 차이가 있음
Post
pTable = Post$p.value
pTableComplete = fullPTable(pTable)
multcompLetters(pTableComplete)
```
